```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(file)  # syntax for reading the file

# Assuming you have the data in a DataFrame named 'df'
# Replace 'ratings_column' and 'cost_column' with the actual column names in your DataFrame

# Sample DataFrame creation for illustration (Replace this with your actual data loading step)
# df = pd.read_csv('your_data_file.csv')

# Creating distplots for the two columns
plt.figure(figsize=(12, 6))

# Distribution of Ratings
plt.subplot(1, 2, 1)
sns.histplot(df['Fare'].dropna(), kde=True, bins=30)
plt.title("Fare price")
plt.xlabel('Price')
plt.ylabel('Frequency')



plt.tight_layout()
plt.show()
```

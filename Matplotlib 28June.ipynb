```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {'Values': np.random.normal(loc=50, scale=15, size=(1000*600))}
df = pd.DataFrame(data)

# Descriptive Statistics
mean = df['Values'].mean()
median = df['Values'].median()
mode = df['Values'].mode()[0]
std_dev = df['Values'].std()
variance = df['Values'].var()
range_ = df['Values'].max() - df['Values'].min()
iqr = df['Values'].quantile(0.75) - df['Values'].quantile(0.25)

print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Mode: {mode}')
print(f'Standard Deviation: {std_dev}')
print(f'Variance: {variance}')
print(f'Range: {range_}')
print(f'IQR: {iqr}')

# Frequency Distribution
plt.figure(figsize=(10, 6))
#sns.histplot(df['Embarked'], kde=True)
plt.title('Histogram with Density Plot')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

# Box Plot
plt.figure(figsize=(8, 4))
#sns.boxplot(x=df['Values'])
plt.title('Box Plot')
plt.xlabel('Values')
plt.show()
```
